```python
# Acessando as séries anuais (todos os dias úteis do ano escolhido)

files_b3 = [] 
for year in range(year_initial,year_final):
  files_b3.append(f'COTAHIST_A{year}.ZIP')
files_b3
```

```python
# Definindo o nome e largura das colunas dos arquivos importados da B3

# http://www.b3.com.br/data/files/C8/F3/08/B4/297BE410F816C9E492D828A8/SeriesHistoricas_Layout.pdf

names_cols = ['TIPREG',# TIPO DE REGISTRO,DATA DO PREGÃO
              'DATPRE',# DATA DO PREGÃO
              'CODBDI',# CÓDIGO BDI
              'CODNEG',# CÓDIGO DE NEGOCIAÇÃO DO PAPEL
              'TPMERC',# TIPO DE MERCADO
              'NOMRES',# NOME RESUMIDO DA EMPRESA EMISSORA DO PAPEL
              'ESPECI',# ESPECIFICAÇÃO DO PAPEL
              'PRAZOT',# PRAZO EM DIAS DO MERCADO A TERMO
              'MODREF',# MOEDA DE REFERÊNCIA
              'PREABE',# PREÇO DE ABERTURA DO PAPEL - MERCADO NO PREGÃO
              'PREMAX',# PREÇO MÁXIMO DO PAPEL - MERCADO NO PREGÃO
              'PREMIN',# PREÇO MÍNIMO DO PAPEL- MERCADO NO PREGÃO
              'PREMED',# PREÇO MÉDIO DO PAPEL- MERCADO NO PREGÃO
              'PREULT',# PREÇO DO ÚLTIMO NEGÓCIO DO PAPEL - MERCADO NO PREGÃO
              'PREOFC',# PREÇO DA MELHOR OFERTA DE COMPRA DO PAPELMERCADO
              'PREOFV',# PREÇO DA MELHOR OFERTA DE VENDA DO PAPELMERCADO
              'TOTNEG',# NEG. - NÚMERO DE NEGÓCIOS EFETUADOS COM O PAPEL- MERCADO NO PREGÃO
              'QUATOT',# QUANTIDADE TOTAL DE TÍTULOS NEGOCIADOS NESTE PAPEL- MERCADO
              'VOLTOT',# VOLUME TOTAL DE TÍTULOS NEGOCIADOS NESTE PAPEL- MERCADO
              'PREEXE',# PREÇO DE EXERCÍCIO PARA O MERCADO DE OPÇÕES OU VALOR DO CONTRATO PARA O MERCADO DE TERMO SECUNDÁRIO
              'INDOPC',# INDICADOR DE CORREÇÃO DE PREÇOS DE EXERCÍCIOS OU VALORES DE CONTRATO PARA OS MERCADOS DE OPÇÕES OU TERMO SECUNDÁRIO
              'DATVEN',# DATA DO VENCIMENTO PARA OS MERCADOS DE OPÇÕES OU TERMO SECUNDÁRIO
              'FATCOT',# FATOR DE COTAÇÃO DO PAPEL
              'PTOEXE',# PREÇO DE EXERCÍCIO EM PONTOS PARA OPÇÕES REFERENCIADAS EM DÓLAR OU VALOR DE CONTRATO EM PONTOS PARA TERMO SECUNDÁRIO
              'CODISI',# CÓDIGO DO PAPEL NO SISTEMA ISIN OU CÓDIGO INTERNO DO PAPEL
              'DISMES',# NÚMERO DE DISTRIBUIÇÃO DO PAPEL
]

widths_cols = [2, 8, 2, 12, 3, 12, 10, 3, 4, 13, 13, 13, 13, 13, 13, 13, 5, 18, 18, 13, 1, 8, 7, 13, 12, 3]
```

```python
# Função para realizar o pré-processamento (ajuste de layout) dos arquivos importados da B3

def process(file_path):

  # Função para carregar o arquivo bruto (não considera o HEADER e TRAILER), 
  df0 = pd.read_fwf(file_path, widths=widths_cols, names=names_cols, skiprows=1, skipfooter=1)

  # Convertendo as colunas 'PREABE', 'PREMAX', 'PREMIN', 'PREMED', 'PREULT', 'PREOFC', 'PREOFV', 'PREEXE', 'PTOEXE' para tipo float
  df0[['PREABE', 'PREMAX', 'PREMIN', 'PREMED', 'PREULT', 'PREOFC', 'PREOFV', 'PREEXE', 'PTOEXE']] = df0[['PREABE', 'PREMAX', 'PREMIN', 'PREMED', 'PREULT', 'PREOFC', 'PREOFV', 'PREEXE', 'PTOEXE']].astype('float64')
  
  # Convertendo as colunas 'DATPRE', 'CODBDI', 'CODNEG', 'NOMRES', 'ESPECI', 'PRAZOT', 'MODREF', 'DATVEN', 'CODISI' para tipo string
  df0[['DATPRE', 'CODBDI', 'CODNEG', 'NOMRES', 'ESPECI', 'PRAZOT', 'MODREF', 'DATVEN', 'CODISI']] = df0[['DATPRE', 'CODBDI', 'CODNEG', 'NOMRES', 'ESPECI', 'PRAZOT', 'MODREF', 'DATVEN', 'CODISI']].astype('string')
  
  # Convertendo as colunas 'TIPREG', 'TPMERC', 'TOTNEG', 'QUATOT', 'VOLTOT', 'INDOPC', 'FATCOT', 'DISMES' para tipo int64
  df0[['TIPREG', 'TPMERC', 'TOTNEG', 'QUATOT', 'VOLTOT', 'INDOPC', 'FATCOT', 'DISMES']] = df0[['TIPREG', 'TPMERC', 'TOTNEG', 'QUATOT', 'VOLTOT', 'INDOPC', 'FATCOT', 'DISMES']].astype('int64')
  
  # Dividindo as colunas PREABE, PREMAX, PREMIN, PREMED, PREULT, PREOFC, PREOFV por 100 (ou multiplicar por 0.01)
  df0[['PREABE', 'PREMAX', 'PREMIN', 'PREMED', 'PREULT', 'PREOFC', 'PREOFV', 'PREEXE', 'PTOEXE']] = df0[['PREABE', 'PREMAX', 'PREMIN', 'PREMED', 'PREULT', 'PREOFC', 'PREOFV', 'PREEXE', 'PTOEXE']]*0.01
  
  # Convertendo a coluna 'DATPRE' para o formato AAAA-MM-DD
  df0['DATPRE'] = df0['DATPRE'].apply(lambda x: datetime.strptime(x, '%Y%m%d'))

  return(df0)  
```

```python
# Definindo a lista de arquivos TXT a serem pré-processados

files_b3 = []
for year in range(year_initial,year_final):
  files_b3.append(f'COTAHIST_A{year}.TXT')
files_b3
```

```python
# Realizando o pré-processamento dos arquivos e salvando como csv

for file in files_b3:
  df = process(b3_path+file)
  name = file.strip(".TXT") # exclui a extensão TXT do nome do arquivo
  df.to_csv(b3_path+name+'.CSV',index=False)
```

```python
# Carregando o dataframe 

df1 = pd.read_csv('Dados\B3\dados.csv',sep='|',encoding='utf-16')
df1.drop('Unnamed: 0',axis=1,inplace=True) # Excluindo coluna desnecessária
df1
```

```python
# Copiando a coluna 'Códigos de Negociação' para criar as colunas 'Códigos ISIN' e 'Códigos CVM'

df = pd.DataFrame(df1['Códigos de Negociação:'])
df
```

```python
# Excluindo a coluna 'Códigos de Negociação' que será desmenbrada em colunas adicionais

df1.drop('Códigos de Negociação:',axis=1,inplace=True)
df1
```

```python
df1.columns
```

```python
# Inserindo o separador '|' para separar as células em colunas adicionais

df['Códigos de Negociação:'] = df['Códigos de Negociação:'].str.replace('Códigos CVM:','|Códigos CVM:')
df['Códigos de Negociação:'] = df['Códigos de Negociação:'].str.replace('Códigos ISIN:','|Códigos ISIN:')
df['Códigos de Negociação:'] = df['Códigos de Negociação:'].str.replace('Nenhum ativo no Mercado a Vista – Lote Padrão.','NENHUM')
df
```

```python
df.columns
```

```python
# Separando a coluna 'Códigos de Negociação' em 3 colunas (Códigos de Negociação, Códigos ISIN e Códigos CVM)

split = df['Códigos de Negociação:'].str.split("|",n=2,expand=True)
split
```

```python
# Criando o dataframe 'Códigos de Negociação'

negociacao = pd.DataFrame(split[0])
negociacao
```

```python
# Fazendo o split da coluna 'Códigos de Negociação'

negociacao = negociacao[0].str.split("Mais Códigos ",expand=True)
negociacao.drop([0],axis=1,inplace=True)
negociacao
```

```python
negociacao[1] = negociacao[1].str.replace(';','') # Substituindo o caracter ';' por 'vazio'
negociacao
```

```python
negociacao[1] = negociacao[1].str.replace(' ',';') # Substituindo o caracter 'vazio' por ';'
negociacao
```

```python
negociacao[1] = negociacao[1].str.replace(';;',';') # Substituindo o caracter ';;' por ';'
negociacao
```

```python
negociacao = negociacao[1].str.split(";",n=13,expand=True) # O separador é o caracter ';'
negociacao
```

```python
negociacao.drop([0],axis=1,inplace=True)
negociacao
```

```python
negociacao.rename(columns = {1:'Códigos de Negociação1:',2:'Códigos de Negociação2:',3:'Códigos de Negociação3:',4:'Códigos de Negociação4:',5:'Códigos de Negociação5:',6:'Códigos de Negociação6:',7:'Códigos de Negociação7:',8:'Códigos de Negociação8:',9:'Códigos de Negociação9:',10:'Códigos de Negociação10:',11:'Códigos de Negociação11:',12:'Códigos de Negociação12:',13:'Códigos de Negociação13:'}, inplace = True)
negociacao
```

```python
# Criando o dataframe 'Códigos ISIN'

isin = pd.DataFrame(split[1])
isin
```

```python
# Fazendo o split da coluna 'Códigos ISIN'

isin = isin[1].str.split("Códigos ISIN: ",expand=True)
isin.drop([0],axis=1,inplace=True)
isin
```

```python
isin = isin[1].str.split(",",n=10,expand=True) # O separador é o caracter ','
isin
```

```python
isin.rename(columns = {0:'Códigos ISIN1:',1:'Códigos ISIN2:',2:'Códigos ISIN3:',3:'Códigos ISIN4:',4:'Códigos ISIN5:',5:'Códigos ISIN6:',6:'Códigos ISIN7:',7:'Códigos ISIN8:'}, inplace = True)
isin
```
